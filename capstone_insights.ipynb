```python
countscategory = df['category'].value_counts() # find the counts for each unique category
countscategory
```

```python
#Distribution of categories and what trended the most combined for all countries.



labels = ['Entertainment', 'People & Blogs', 'News & Politics', 'Sports', 'Comedy', 'Music', 'Howto & Style', 'Film & Animation',
        'Gaming', 'Education', 'Science and Technolgy', 'Auto & Vehicles', 'Pets & Animals', 'Nonprofits & Activism', 'Travel & Events',
         'Shows', 'Movies', 'Trailers']
sizes = [50468, 28045, 20327, 12561, 10358, 10145, 9580, 8460, 5318, 3897, 3404, 3274, 2271, 1779, 812, 435, 12, 3]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral', 'red', 'orange', 'purple', 'grey', 'pink', 'teal', 'oldlace', 'brown', 'limegreen', 'maroon', 'rosybrown', 'navy', 'violet', 'cadetblue']

patches, texts, pct = plt.pie(sizes, colors=colors, autopct = '%1.1f%%', pctdistance=1.06, labeldistance=1.1, shadow=True, startangle=90)
plt.legend(patches, labels, loc="best")
plt.axis('equal')
plt.tight_layout()
fig = plt.gcf()
fig.set_size_inches(12,12)
plt.show()

```

```python
#the count of trending videos in each category. #same info but different visualization of pie chart above

fig = plt.figure(figsize=(10,10)) # define plot area
ax = fig.gca() # define axis    
countscategory.plot.bar(ax = ax) # Use the plot.bar method on the counts data frame
ax.set_title('Count of categories') # Give the plot a main title
ax.set_xlabel('Categories') # Set text for the x axis
ax.set_ylabel('Count')# Set text for y axis
```
